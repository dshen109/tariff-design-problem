```python
# Make queries print out the bash command to run them

import os
import sys
import yaml
import numpy as np

OPTIMISTIC = 25
PESSIMISTIC = 25

MIN_PROFIT = 10
MIN_APPROVAL = 10
CARBON_CREDIT_BUDGET = 30
MAX_BILL = 150

QUERY_FOLDER = "/home/dshen/Documents/category-theory/final-proj/tariff-design-problem/tariff_design_v2.mcdplib/queries"

OPTIMIZE_FOR_FixFunMinRes = ["max_bill", "carbon_credit_budget"]
OPTIMIZE_FOR_FixResMaxFun = ["approval", 'profit']


class FixFunMinRes:
    def __init__(self, title, description="", model="`tariff_design"):
        self.title = title
        self.description = description
        self.model = model

    def gen_yaml(self, profit, approval, carbon_credit_budget, max_bill, optimize_for):
        output = {
            "title": self.title,
            "description": self.description,
            "model": self.model,
            "query": {
                "query_type": "FixFunMinRes",
                "min_f": {},
                "max_r": {},
                "optimize_for": [optimize_for] if isinstance(optimize_for, str) else optimize_for
            }
        }
        output['query']['min_f']["profit"] = f"{profit} USD"
        output['query']['min_f']["approval"] = f"{approval} dimensionless"
        output['query']['max_r']["carbon_credit_budget"] = f"{carbon_credit_budget} USD"
        output['query']['max_r']["max_bill"] = f"{max_bill} USD"
        return output


class FixResMaxFun:
    def __init__(self, title, description="", model="`tariff_design"):
        self.title = title
        self.description = description
        self.model = model

    def gen_yaml(self, profit, approval, carbon_credit_budget, max_bill, optimize_for):
        output = {
            "title": self.title,
            "description": self.description,
            "model": self.model,
            "query": {
                "query_type": "FixResMaxFun",
                "min_f": {},
                "max_r": {},
                "optimize_for": [optimize_for] if isinstance(optimize_for, str) else optimize_for
            }
        }
        output['query']['min_f']["profit"] = f"{profit} USD"
        output['query']['min_f']["approval"] = f"{approval} dimensionless"
        output['query']['max_r']["carbon_credit_budget"] = f"{carbon_credit_budget} USD"
        output['query']['max_r']["max_bill"] = f"{max_bill} USD"
        return output


# Tighten the approval requirement
# for approval in np.arange(OPTIMIZE_FOR_FixFunMinRes, 50, 10):
#     continue
#     optimize_for = '_'.join(OPTIMIZE_FOR)
#     fname = f"fixfunminres_approval_{approval}_co2_{CARBON_CREDIT_BUDGET}_maxbill_{MAX_BILL}_profit_{MIN_PROFIT}_optimizefor_{optimize_for}"
#     if not os.path.exists(QUERY_FOLDER):
#         os.makedirs(QUERY_FOLDER)
#     query = FixFunMinRes(fname)
#     yaml_output = query.gen_yaml(MIN_PROFIT, approval, CARBON_CREDIT_BUDGET,
#                                  MAX_BILL, OPTIMIZE_FOR)
#     # Write the yaml file
#     with open(f"{QUERY_FOLDER}/{fname}.mcdp_query.yaml", "w") as f:
#         yaml.dump(yaml_output, f)

#     print(
#         f"mcdp-solve-query  --optimistic {OPTIMISTIC} "
#         f"--pessimistic {PESSIMISTIC} "
#         f"{fname} --imp --out queries/{fname}"
#     )


# Tighten the bill constraint
for max_bill in np.arange(MAX_BILL, 50, -30):
    optimize_for = '_'.join(OPTIMIZE_FOR_FixResMaxFun)
    required_profit = 30
    fname = f"fixresmaxfun_approval_{MIN_APPROVAL}_co2_{CARBON_CREDIT_BUDGET}_maxbill_{max_bill}_profit_{required_profit}_optimizefor_{optimize_for}"
    if not os.path.exists(QUERY_FOLDER):
        os.makedirs(QUERY_FOLDER)
    query = FixResMaxFun(fname)
    yaml_output = query.gen_yaml(required_profit, MIN_APPROVAL, CARBON_CREDIT_BUDGET,
                                 max_bill, OPTIMIZE_FOR_FixResMaxFun)
    # Write the yaml file
    with open(f"{QUERY_FOLDER}/{fname}.mcdp_query.yaml", "w") as f:
        yaml.dump(yaml_output, f)

    print(
        f"mcdp-solve-query  --optimistic {OPTIMISTIC} "
        f"--pessimistic {PESSIMISTIC} "
        f"{fname} --imp --out queries/{fname}"
    )

```
